```python
# Ejercicio 4
def palindroma(word):
    if word == str(word):
        letters = []
        for e in word:
            letters.append(e)
        n = len(letters)//2
        for i in range(n):
            if letters[i] != letters[-1*(i+1)]:
                return False
            else:
                continue
        return True

palindroma = np.frompyfunc(palindroma,1 ,1)

words = np.array([["ala", "delfín", "arroz"], ["cena", "kayak", "picnic"], ["hoja", "gato", "elle"]])
print(palindroma(words))
```

```python
# Ejercicio 7

def ingresar():
    nums = []
    while len(nums) != 5:
        n = int(input('Ingresa un número entre 1 y 20: '))
        if n >= 1 and n <= 20:
            nums.append(n)
        else:
            continue
    nums = np.array(nums)
    return nums

def metodo_ac(a, nums):
    ac = 0
    for e in a:
        if e in nums:
            ac += 1
        else:
            continue
    return ac

def metodo_pos_ac(a, nums):
    ac = 0
    pos_ac = 0
    for i in range(5):
        for j in range(len(nums)):
            if a[i] == nums[j]:
                ac += 1
                if  i == j:
                    pos_ac +=1
                else:
                    continue
            else:
                continue
    return pos_ac

a = np.random.randint(1, 20, size =5)
ac = 0
pos_ac = 0

while ac != 5 and pos_ac != 5:
    nums = ingresar()
    ac = metodo_ac(a, nums)
    pos_ac = metodo_pos_ac(a, nums)
    print('Has acertado {} números del array y {} posiciones'.format(ac, pos_ac))
    print(nums,'\n')
print('¡¡¡ENHORABUENA!!! Has acertado el array que había pensado: {}'.format(a))
```

```python
# Ejercicio 9

def vocales(words):
    words = words.lower()
    vowels = 0
    for i in range(len(words)):
        if (words[i] == "a" or words[i] =="e" or words[i] == "i" or words[i] == "o" or words[i] == "u"):
            vowels += 1
        else:
            continue
    return vowels

vowels = np.frompyfunc(vocales,1 , 1)
strings = np.array(['mariposa','flor', 'ada', 'libro', 'amanecer', 'gato', 'razonar'])

vowels(strings)
```

```python
# Ejercicio 10

def str_into_set(string, conjunto):
    if string in conjunto:
        return True
    else:
        return False
       
str_into_set = np.frompyfunc(str_into_set, 2, 1)

strings = np.array(['a','b','c','d','e','f','g','h','i'])
conjunto = {'a','e','i','o','u'}

str_into_set(strings, conjunto)
```
