```python
#Ejercicio 3
def num_letters(words):
    words = words.lower()
    vowels = 0
    consonants = 0
    for i in range(len(words)):
        if (words[i] == "a" or words[i] =="e" or words[i] == "i" or words[i] == "o" or words[i] == "u"):
            vowels += 1
        elif words[i] == str(words[i]):
            consonants += 1
        else:
            continue
    if vowels > consonants:
        return words
```

```python
word = ["miniserie","television","temario","escritorio","dibujo"]
list(filter(num_letters, word))
```

```python
#Ejercicio 4
def divisor(n):
    div = []
    for i in range(1, n+1):
        if n % i == 0:
            div.append(i)
        else:
            continue
    if len(div)>= 5:
        return n
```

```python
numbers = [10, 24, 50, 21, 40, 12, 9]
list(filter(divisor, numbers))
```

```python
#Ejercicio 5
def longitud(word1, word2):
    if len(word1) > len(word2):
        return word1
    else:
        len(word1) < len(word2)
        return word2
```

```python
from functools import reduce
word = ["pelota","juguete", "avion","coche", "patineta", "artista", "pajarito"]
reduce(longitud, word)
```

```python
#Ejercicio 6
def vocales(words):
    words = words.lower()
    vowels = 0
    for i in range(len(words)):
        if (words[i] == "a" or words[i] =="e" or words[i] == "i" or words[i] == "o" or words[i] == "u"):
            vowels += 1
        else:
            continue
    return vowels
```

```python
words = ["manzana", "pera", "mango", "uva", "platano", "aguacate"]
list(map(vocales, words))
```

```python
#Ejercicio 7
def consonantes(words):
    words = words.lower()
    consonants = 0
    for i in range(len(words)):
        if (words[i] == "a" or words[i] =="e" or words[i] == "i" or words[i] == "o" or words[i] == "u"):
            continue
        else:
            consonants += 1
    return consonants
```

```python
def max_or_min(a, b):
    if a > b:
        return a
    else:
        return b
```

```python
from functools import reduce
words = ["papaya", "pera", "mango", "uva", "platano", "aguacate"]
x = list(map(consonantes, words))
y = reduce(max_or_min, x)
words[x.index(y)]
```
