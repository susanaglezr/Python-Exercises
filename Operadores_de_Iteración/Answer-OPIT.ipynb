```python
#Ejercicio 1
num = int(input("Introduce un número entero y se mostrará si es par o impar: "))
while num != 0:
    nm = num % 2 
    if nm == 0:
        print("El número es par")
    else:
        print("El número es impar") 
    num = int(input("Introduce un número entero y se mostrará si es par o impar: "))
```

```python
#Ejercicio 2
word = input("Introduce una palabra: ")
let = input("Introduce una letra: ")
word.lower()
let.lower()
for c in word:
    if c == let:
        print("La palabra contiene la letra dada")
        break
else:
    print("La palabra no contiene la letra dada")
    
```

```python
#Ejercicio 3
num = int(input("Introduce el precio: "))
sum = num
while (sum<= 200):
    num = int(input("Introduce el precio: "))
    sum += num
    if num == 0:
        print("El precio final es:",sum)
        break
else:
    print("Se ha pasado del presupuesto de 200 euros, usted introdujo:", sum)
```

```python
#Ejercicio 4
num = int(input("Introduce un número entero positivo o negativo: "))
sum_1 = 0
sum_2 = 0
while num != 0 :
    if num > 0:
        sum_1 += 1
    else:
        sum_2 += 1
    num = int(input("Introduce un número entero positivo o negativo: "))
print("Ingresaste {} números positivos y {} números negativos".format(sum_1, sum_2))
```

```python
#Ejercicio 5
num = float(input("Introduce un número (se dará la media aritmética al finalizar el programa): "))
sum = 0
count = 0
while num != 0:
    sum += num
    num = float(input("Introduce otro número: "))
    count += 1
med = sum / count
print("La media aritmética de los números dados es:",med)
```

```python
#Ejercicio 6
num_1 = int(input("Introduce el número del extremo izquierdo del intervalo: "))
num_2 = int(input("Introduce el número del extremo derecho del intervalo: "))
print("Los números en el intervalo son:")
for i in range(num_1, num_2 + 1):
    print(i)
```

```python
#Ejercicio 7
num_1 = int(input("Introduce el número del extremo izquierdo del intervalo: "))
num_2 = int(input("Introduce el número del extremo derecho del intervalo: "))
sum = 0
for i in range(num_1, num_2 + 1):
    if (i % 3 == 0):
        sum += i
    else: 
        continue
print("La suma de los números múltiplos de 3 en el intervalo dado es:",sum)
```

```python
#Ejercicio 8
cant = int(input("¿Cuántos números va a introducir?: "))
mult = 1
for i in range(cant):
    num = float(input("Ingresa un número: "))
    mult *= num
print("El producto de todos los números es:",mult)
    
```

```python
#Ejercicio 9
edad = int(input("Ingresa tu edad: "))
year = int(input("Ingresa el año actual: "))
nacim = year - edad
cumple = input("¿Ya cumpliste años en el año actual?, responde si o no: ")
cumple.lower()
if (cumple == "si"):
    for i in range(nacim, year + 1):
        print(i)
else:
    for i in range(nacim - 1, year + 1):
        print(i)
```
