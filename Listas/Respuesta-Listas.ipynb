```python
#Ejercicio 2
string_1 = str(input("Ingrese el primer string: "))
string_2 = str(input("Ingrese el segundo string: "))

lista1 = list(string_1)
lista2 = list(string_2)
lista_final = []

if (len(string_1) == len(string_2)):
    largo = len(string_1)
    for i in range(largo):
        lista3 = lista1[i] + lista2[i]
        lista_final += lista3
    print(list(lista_final))
else:
    print("Los strings no son de la misma longitud")
```

```python
#Ejercicio 3
string_1 = str(input("Introduce un string y se guardaran las consonantes en una lista: "))
string_1.lower()
lista = []
largo = len(string_1)
for i in range(largo):
    letra = string_1[i]
    if letra != "a" and letra != "e" and letra != "i" and letra != "o" and letra != "u":
        sub_lista = list(letra)
        lista += sub_lista
print(lista)
```

```python
#Ejercicio 7
A = []
sum = 0
B = []
n = int(input("Ingrese el número de filas: "))
m = int(input("Ingrese el número de columnas: "))
print("")
for i in range(n):
    A.append([])
    for j in range(m):
        A[i].append(float(input("Ingrese el valor en la posición ({}, {}): ".format(i, j))))
print("")
print("La matriz ingresada es: ")
for i in range(n):
    for j in range(m):
        print(A[i][j], end = "  " if (A[i][j] >= 0) and (A[i][j] < 10) else " ")
    print("")

for j in range(m):
    for i in range(n):
        sum += A[i][j]
    B.append([sum])
    sum = 0
print("")
for i in range(n-1):
    if B[i] != B[i+1]:
        print("La suma de las columnas {} y {} son diferentes".format(i, i+1))
    else:
        print("La suma de las columnas {} y {} son iguales".format(i, i+1))
print(B)
```

```python
#Ejercicio 10 
import numpy as np
n = int(input("Ingrese la dimensión de la matriz: "))

A = np.empty((n, n))
for i in range(n):
    for j in range(n):
        if i == j:
            A[i][j] = 1
        else:
            A[i][j]=0
print(A)
```
