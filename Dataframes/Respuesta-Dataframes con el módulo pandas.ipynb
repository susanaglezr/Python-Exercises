```python
# Ejercicio 6
word = ["euro", "diez", "algas", "broma", "cicuta", "fatiga", "nachos", "jadeos", "hazañas", "boutique"]
vowels = []
consonants = []
length = []
for e in word:
    length.append(len(e))
    vowel = 0
    cons = 0
    for i in range(len(e)):
        if e[i] == "a" or e[i] == "e" or e[i] == "i" or e[i] == "o" or e[i] == "u":
            vowel += 1
        else:
            cons +=1 
    vowels.append(vowel)
    consonants.append(cons)
data = {"WORD": word}
words = pd.DataFrame(data)
words.insert(loc = 1, column = "LENGTH", value = length)
words.insert(loc = 2, column = "VOWELS", value = vowels)
words.insert(loc = 3, column = "CONSONANTS", value = consonants)
words
```

```python
# Ejercicio 7
for i in range(len(words)):
    if words.iloc[i, 2] == 1:
        print(f"La palabra {words.iloc[i,0]} tiene {words.iloc[i,1]} letras de las cuales {words.iloc[i,2]} es vocal y {words.iloc[i,3]} consonantes.")
    elif words.iloc[i, 3] == 1:
        print(f"La palabra {words.iloc[i,0]} tiene {words.iloc[i,1]} letras de las cuales {words.iloc[i,2]} son vocales y {words.iloc[i,3]} consonante.")
    else:
        print(f"La palabra {words.iloc[i,0]} tiene {words.iloc[i,1]} letras de las cuales {words.iloc[i,2]} son vocales y {words.iloc[i,3]} consonantes.")
```
