```python
# Ejercicio 10

def addzero(num, digits):
    num = str(num)
    zeros = digits - len(num)
    if zeros != 0:
        for i in range(zeros):
            num =  '0'+ num 
        return num
    return num

class Date(object):
    def __init__(self, day = 1, month = 1, year = 1):
        self.day = day
        self.month = month
        self.year = year
        if self.validDate() == False:
            print('¡¡¡La fecha introducida no es una fecha válida!!!') 

    def __str__(self):
        return '{} / {} / {}'.format(addzero(self.day, 2), addzero(self.month, 2), addzero(self.year, 4))

    def isLeap(self):
        if self.year % 4 == 0:
            if self.year % 100 == 0:
                if self.year % 400 == 0:
                    return True
                return False
            return True
        return False

    def totalMonthDays(self):
        if self.month < 8:
            if self.month != 2:
                for i in range(1, 8):
                    if self.month % 2 == 0:
                        return 30
                    return 31
            if self.isLeap() == True:
                return 29
            return 28
        for i in range(8, 13):
            if self.month % 2 == 0:
                return 31
            return 30

    def validDate(self):
        if self.day > 0 and self.day <= self.totalMonthDays():
            if self.month > 0 and self.month <= 12:
                if self.year == int(self.year):
                    return True
                return False
            return False
        return False

    @property
    def monthName(self):
        months = {1: 'January', 2: 'February', 3: 'March', 4: 'April',
                  5: 'May', 6: 'June', 7: 'July', 8: 'August',
                  9: 'September', 10: 'October', 11: 'November',
                  12: 'December'}
        return months[self.month]

    @staticmethod
    def areEqual(date1, date2):
        if date1.day == date2.day and date1.month == date2.month and date1.year == date2.year:
            return True
        return False

    @staticmethod
    def isLater(date1, date2):
        if date1.year == date2.year:
            if date1.month == date2.month:
                if date1.day > date2.day:
                    return True
                return False
            elif date1.month > date2.month:
                return True
            else:
                return False
        elif date1.year > date2.year:
            return True
        else:
            return False

    @staticmethod
    def isPrevious(date1, date2):
        if date1.year == date2.year:
            if date1.month == date2.month:
                if date1.day < date2.day:
                    return True
                return False
            elif date1.month < date2.month:
                return True
            else:
                return False
        elif date1.year < date2.year:
            return True
        else:
            return False

    @classmethod
    def firstDayOfTheYear(cls, year):
        day = 1
        month = 1
        return cls(day, month, year)

    @classmethod
    def lastDayOfTheYear(cls, year):
        day = 31
        month = 12 
        return cls(day, month, year)

    @classmethod
    def plusDay(cls, date):
        if date.day == date.totalMonthDays() and (date.month != 12):
            return cls(1, date.month + 1, date.year)
        elif (date.day == 31) and (date.month == 12):
            return cls(1, 1, date.year + 1)
        else:
            return cls(date.day + 1, date.month, date.year)

    @classmethod
    def minusDay(cls, date):
        if (date.day == 1) and (date.month != 1):
            date.month = date.month - 1
            dateMonth = Date(date.day, date.month, date.year)
            date.day = dateMonth.totalMonthDays()
            return cls(date.day, date.month, date.year)
        elif (date.day == 1) and (date.month == 1):
            return cls(31, 12, date.year - 1)
        else:
            return cls(date.day -1, date.month, date.year)

    @classmethod
    def copy(cls, date):
        day2 = date.day
        month2 = date.month
        year2 = date.year
        return cls(day2, month2, year2)

    @staticmethod
    def difference(date1, date2):
        days = 0
        daysYear = 0
        if Date.isPrevious(date1, date2) == True:
            if date1.year == date2.year:
                if date1.month == date2.month:
                    if date1.day == date2.day:
                        return 0
                    return date2.day - date1.day
                else:
                    it = date2.month - date1.month
                    for i in range(it):
                        date1.month += i
                        monthDays = date1.totalMonthDays()
                        days += monthDays
                    if date1.day != date2.day:
                        difDays = abs(date1.day - date2.day)
                    days = days + difDays
                    return days
            else:
                it = date2.year - date1.year
                for i in range(it):
                    date1.year += i
                    if date1.isLeap() == True:
                        daysYear += 366
                    else:
                        daysYear += 365    
                it = date2.month - date1.month
                if it != 0:
                    for j in range(it):
                        date1.month += j
                        monthDays = date1.totalMonthDays()
                        days += monthDays
                    if date1.day != date2.day:
                        difDays = abs(date1.day - date2.day)
                        days = days + difDays + daysYear
                        return days 
                else:
                    if date1.day != date2.day:
                        difDays = abs(date1.day - date2.day)
                        days = days + difDays + daysYear
                        return days 

        elif Date.isPrevious(date1, date2) == False:
            if date2.year == date1.year:
                if date2.month == date1.month:
                    if date2.day == date1.day:
                        return 0
                    return date1.day - date2.day
                else:
                    it = date1.month - date2.month
                    for i in range(it):
                        date2.month += i
                        monthDays = date2.totalMonthDays()
                        days += monthDays
                    if date2.day != date1.day:
                        difDays = abs(date2.day - date1.day)
                    days = days + difDays
                    return days
            else:
                it = date1.year - date2.year
                for i in range(it):
                    date2.year += i
                    if date2.isLeap() == True:
                        daysYear += 366
                    else:
                        daysYear += 365    
                it = date1.month - date2.month
                if it != 0:
                    for j in range(it):
                        date2.month += j
                        monthDays = date2.totalMonthDays()
                        days += monthDays
                    if date2.day != date1.day:
                        difDays = abs(date2.day - date1.day)
                        days = days + difDays + daysYear
                        return days 
                else:
                    if date2.day != date1.day:
                        difDays = abs(date2.day - date1.day)
                        days = days + difDays + daysYear
                        return days
        else:
            return 0      


    @classmethod
    def randomDate(cls):
        import random
        day = random.randrange(1, 32)
        month = random.randrange(1, 13)
        year = random.randrange(1, 3000)
        return cls(day, month, year)

    @classmethod
    def toDate(cls, date):
        date = date.split(' ')
        day = int(date[0])
        dicMonths = {'January':1, 'February':2,	'March':3, 'April':4, 'May':5, 'June':6, 
                     'July':7, 'August':8, 'September':9, 'October':10,
                     'November':11, 'December':12}
        month = dicMonths[date[1]]
        year = int(date[2])
        return cls(day, month, year) 
```
