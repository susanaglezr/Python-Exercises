```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Ejercicio 1
x = np.random.randint(100, size = 100)
y = np.random.randint(100, size = 100)
colors = np.random.randint(100, size = 100)
sizes = 10*(np.random.randint(100, size = 100))

plt.figure(figsize = (10,10))
plt.scatter(x, y, alpha = 0.5, s = sizes, c = colors, cmap = 'hsv')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
plt.show()
```

```python
# Ejercicio 2
real = [1, 0, -1, -3, -2, 0, 1]
im = [1, 2, -1, 0, -1, -2, -1]

points = 100
plt.scatter(real, im, c = 'red', s = 120)

l_eje1 = np.linspace(-3.2, 3.2, points)
l_eje2 = np.zeros(points)
plt.plot(l_eje1, l_eje2, c = 'k', alpha = 0.5)
plt.plot(l_eje2, l_eje1,c = 'k', alpha = 0.5)

plt.title('Plano complejo')
plt.xlabel('Eje real')
plt.ylabel('Eje imaginario')

plt.xlim([-3.2, 3.2])
plt.ylim([-2.2, 2.2])

plt.show()
```

```python
# Ejercicio 4
def f1(x):
    resultado = math.sin(x)/x
    return resultado

def f2(x):
    resultado = (2*math.sin(x))/x
    return resultado

def f3(x):
    resultado = -math.sin(x)/x
    return resultado

x = np.linspace(-5*math.pi, 5*math.pi, 100)

y1 = [f1(i) for i in x]
y2 = [f2(i) for i in x]
y3 = [f3(i) for i in x]

plt.figure(figsize = (8,6))
plt.plot(x, y1, c = '#f06741')
plt.plot(x, y2, c = '#a4f041')
plt.plot(x, y3, c = '#8d41f0')

plt.show()
```
