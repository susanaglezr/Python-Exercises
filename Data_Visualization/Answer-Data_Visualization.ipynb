```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Ejercicio 1
x = np.random.randint(100, size = 100)
y = np.random.randint(100, size = 100)
colors = np.random.randint(100, size = 100)
sizes = 10*(np.random.randint(100, size = 100))

plt.figure(figsize = (10,10))
plt.scatter(x, y, alpha = 0.5, s = sizes, c = colors, cmap = 'hsv')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
plt.show()
```

```python
# Ejercicio 2
real = [1, 0, -1, -3, -2, 0, 1]
im = [1, 2, -1, 0, -1, -2, -1]

points = 100
plt.scatter(real, im, c = 'red', s = 120)

l_eje1 = np.linspace(-3.2, 3.2, points)
l_eje2 = np.zeros(points)
plt.plot(l_eje1, l_eje2, c = 'k', alpha = 0.5)
plt.plot(l_eje2, l_eje1,c = 'k', alpha = 0.5)

plt.title('Plano complejo')
plt.xlabel('Eje real')
plt.ylabel('Eje imaginario')

plt.xlim([-3.2, 3.2])
plt.ylim([-2.2, 2.2])

plt.show()
```

```python
# Ejercicio 3
def color(n):
    dicc = {10: 'A',11: 'B',12: 'C',13: 'D',14: 'E',15: 'F'}
    num = []
    resultado = []
    div = n // 16
    resto = n % 16
    num.append(div)
    num.append(resto)
    for i in num:
        if i >= 10:
            h = dicc[i]
            resultado.append(h)
        else:
            resultado.append(i)
    R ="".join(map(str, resultado))
    return R         


def RGB_to_H(r,g,b):
    r = color(r)
    g = color(g)
    b = color(b)
    return '#'+r+g+b
```

```python
# Ejercicio 4
def f1(x):
    resultado = math.sin(x)/x
    return resultado

def f2(x):
    resultado = (2*math.sin(x))/x
    return resultado

def f3(x):
    resultado = -math.sin(x)/x
    return resultado

x = np.linspace(-5*math.pi, 5*math.pi, 100)

y1 = [f1(i) for i in x]
y2 = [f2(i) for i in x]
y3 = [f3(i) for i in x]

plt.figure(figsize = (8,6))
plt.plot(x, y1, c = '#f06741')
plt.plot(x, y2, c = '#a4f041')
plt.plot(x, y3, c = '#8d41f0')

plt.show()
```

```python
# Ejercicio 5

plt.figure(figsize = (10,5))
# plot 1
plt.subplot(3, 1, 1)
plt.plot(x, y1, c = '#f06741')

# plot 2
plt.subplot(3, 1, 2)
plt.plot(x, y2, c = '#a4f041')

# plot 3
plt.subplot(3, 1, 3)
plt.plot(x, y3, c = '#8d41f0')

plt.show()
```

```python
# Ejercicio 7
import networkx as nx
G = nx.Graph()
G.add_nodes_from("abcde")
G.add_weighted_edges_from([("a", "b", 1), ("a", "c", 2), ("a", "d", 2), ("a", "e", 5),
("b", "e", 4), ("c", "d", 3), ("d", "e", 1)])
nx.draw(G, with_labels = True, font_weight = 'bold', font_color = 'white',
        node_color ='#ad80fa', edge_color ='#b1ddf5')
plt.show()
```

```python
# Ejercicio 8
dg = list(G.degree(['a','b','c','d','e']))
sizes = []
for i in range(len(dg)):
    size = dg[i][1]*350
    sizes.append(size)
plt.figure(figsize= (8,6))
G = nx.Graph()
G.add_nodes_from("abcde")
G.add_weighted_edges_from([("a", "b", 1), ("a", "c", 2), ("a", "d", 2), ("a", "e", 5),
("b", "e", 4), ("c", "d", 3), ("d", "e", 1)])
nx.draw(G, with_labels = True, font_weight = 'bold', font_color = 'white',
        node_color ='#ad80fa', edge_color ='#b1ddf5', node_size= sizes)
plt.show()
```
